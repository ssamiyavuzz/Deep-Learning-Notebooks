# Artificial Neural Network (ANN)

Yapay Sinir Alar, insan beyninin bilgi ileme eklini taklit etmeye alan bir makine renimi modelidir. Neuron ad verilen birimlerden oluur ve bu birimler arasndaki balantlar (arlklar) zerinden bilgi akn simle eder.

**Ana Bileenler:**

- **Input Layer:**  Verinin modele girdii ilk katmandr ve her bir input layer, veri setindeki bir feature'a karlk gelir.

- **Hiden Layer:** Giri ve k katmanlar arasnda yer alr; bu katmanlardaki dmler, renme ilemi srasnda modelin rendii feature'lar veya pattern'leri temsil eder.

- **Output Layer:**  Modelin tahmin veya snflandrma sonularn sunar.



**alma Prensibi:**

- Model, giri verilerini alr ve arlklar ile aktivasyon fonksiyonlar araclyla ileyerek k verisini retir.

- renme sreci, gerek k ile beklenen k arasndaki fark (loss) minimize edecek ekilde arlklarn gncellenmesi ilemidir (rnein backpropogation algoritmas kullanlarak).

**Avantajlar:**

- Yksek miktarda ve eitlilikteki veriden karmak desenlerin renilmesinde etkilidir.
