# Artificial Neural Network (ANN)

Yapay Sinir Alar, insan beyninin bilgi ileme eklini taklit etmeye alan bir makine renimi modelidir. Neuron ad verilen birimlerden oluur ve bu birimler arasndaki balantlar (arlklar) zerinden bilgi akn simle eder.

Karmak desenleri ve veri ilikilerini tanma, snflandrma, tahmin etme ve karar verme yetenekleriyle, grevlere otomatik olarak renme ve adapte olma kabiliyetine sahiptir.

**Ana Bileenler:**

- **Input Layer:**  Verinin modele girdii ilk katmandr ve her bir input layer, veri setindeki bir feature'a karlk gelir.

- **Hiden Layer:** Giri ve k katmanlar arasnda yer alr; bu katmanlardaki dmler, renme ilemi srasnda modelin rendii feature'lar veya pattern'leri temsil eder.

- **Output Layer:**  Modelin tahmin veya snflandrma sonularn sunar.



**alma Prensibi:**

- Model, giri verilerini alr ve arlklar ile aktivasyon fonksiyonlar araclyla ileyerek k verisini retir.

- renme sreci, gerek k ile beklenen k arasndaki fark (loss) minimize edecek ekilde arlklarn gncellenmesi ilemidir (rnein backpropogation algoritmas kullanlarak).

**Avantajlar:**

- Yksek miktarda ve eitlilikteki veriden karmak desenlerin renilmesinde etkilidir.

- Feature Engineering ihtiyacn kaldrr, nk model gerekli feature'lar veriden otomatik olarak renebilir.



**Dezavantajlar:**

- Modelin nasl bir sonuca ulatn yorumlamak zordur.
 
- Byk veri setleri ve hesaplama gc gerektirebilir.

### Neuron

ANN'de bir "neuron" ya da "dm", biyolojik bir sinir hcresinden esinlenilmi bir ilem birimidir. Temel ilevi, girdi olarak ald veriler zerinde basit bir matematiksel ilem yapmak ve bir kt retmektir.

**levi:**

Her bir neuron, bir veya birden fazla giriten (nceki katmandan gelen sinyaller) alnan veriyi toplar, bir arlkla arpar ve bir aktivasyon fonksiyonu uygulayarak bir kt retir. Bu kt, sonraki katmana iletilir.

**alma Prensibi:**

- **Girdi Arlklandrmas:** Her neuron, kendisine gelen her bir girdiyi bir arlk ile arpar. Bu arlklar, girdilerin nemini ve etkisini belirler. Arlklar, eitim srecinde srekli olarak gncellenir.

- **Toplam ve Bias Ekleme:** Arlklandrlm girdiler toplanr ve genellikle bir "bias" deeri eklenir. Bias, aktivasyon fonksiyonunun eiini ayarlamak iin kullanlr ve bu da neuronun aktive olup olmamasna yardmc olur.

- **Aktivasyon Fonksiyonu:** Toplam sonuca bir aktivasyon fonksiyonu uygulanr. Bu fonksiyon, neuronun ktsn belirler ve genellikle kty belirli bir arala sktrmak iin kullanlr. rnein, ReLU, Sigmoid, veya Tanh gibi aktivasyon fonksiyonlar vardr.



### Input Layer

Giri katman, Yapay Sinir Alarna dardan verilen verinin ilk olarak sunulduu katmandr. Modelin geri kalanyla herhangi bir ilem yapmadan nce veriyi alr.

**levi:**

Giri katman, modelin eitim veya tahmin srasnda ileyecei ham veriyi temsil eder. Her bir neuron bu katmanda, veri setindeki bir feature'a karlk gelir.

**alma Prensibi:**

Bu katmanda herhangi bir ilem yaplmaz; sadece veri, an daha derin katmanlarna iletilmek zere hazrlanr.



### Hiden Layer

Hiden Layer, giri ve k katmanlar arasnda yer alan katmanlardr. Bu katmanlar dorudan dardan herhangi bir veri almaz ya da kt vermezler; bunun yerine, giri katmanndan aldklar veriyi iler ve kt katmanna iletilmek zere dntrrler.

**levi:**

 Gizli katmanlar, veri zerinde karmak dnmler ve zellik karmlar yapar. Modelin renme kapasitesi, genellikle gizli katmanlarn says ve bykl ile dorudan ilikilidir.
 
**alma Prensibi:**

- **Data Transform:** Hiden Layer'lar, giri katmanndan gelen veriyi alr ve arlklar ve biaslar yardmyla dntrr. Her bir hiden layer, veriyi daha karmak bir hale getirerek, renme srecinde faydal olabilecek yeni feature'larn karlmasna yardmc olur.

- **Derinlik ve renme Kapasitesi:** ANN'nin renme kapasitesi, hiden layer'larn says ve her katmandaki neuron says ile dorudan ilikilidir. Daha fazla katman ve neuron, genellikle daha karmak problemleri zme yetenei anlamna gelir, ancak ayn zamanda overfitting riskini de artrabilir.



### Output Layer

ANN'in son katmandr ve modelin tahminlerini ya da kararlarn temsil eden sonular retir.

**levi:**

Bu katmann her bir neuronu, genellikle bir tahmin snf ya da sonucuyla ilikilidir. Aktivasyon fonksiyonu, modelin zmlemeye alt probleme bal olarak seilir (rnein, softmax aktivasyon fonksiyonu ok snfl snflandrma problemleri iin kullanlr).

**alma Prensibi:**

Output layer, modelin sonularn retir. Her bir neuron, genellikle bir tahmin snf veya deeri ile ilikilendirilir.

**Aktivasyon fonksiyonu, problemin trne bal olarak seilir:**

**Classification iin:** Multiclass problemler iin softmax, Binary iin sigmoid kullanlr.

**Regression iin:** Lineer aktivasyon fonksiyonu, ktnn herhangi bir gerek say olabilecei durumlar iin kullanlr.
